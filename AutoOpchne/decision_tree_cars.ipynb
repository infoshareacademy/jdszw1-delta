```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cut_data.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data = data.drop("Unnamed: 0", axis='columns')
```

```python
data.groupby("make_name").count()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data['product_tier_n'] = le.fit_transform(data['product_tier'])
data['make_name_n'] = le.fit_transform(data['make_name'])
```

```python
data.columns
```

```python
X = data.drop(['product_tier', 'make_name',
       'first_registration_year', 'created_date', 'deleted_date',
        'detail_views', 'ctr'] ,axis='columns')
yd = data.detail_views
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, yd_train, yd_test = train_test_split(X, yd, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, yd_train)
```

```python
yd_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(yd_test, yd_pred))
print('Mean Squared Error:', metrics.mean_squared_error(yd_test, yd_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(yd_test, yd_pred)))
```

```python
regressor.score(X_test, yd_test)
```
